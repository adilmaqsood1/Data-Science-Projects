# visualization

```python
# 1. Count Plot of Passengers by Class (Pclass)
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Count of Passengers by Class')
plt.xlabel('Class (Pclass)')
plt.ylabel('Count')
plt.show()

# 2. Distribution of Passenger Ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Count Plot of Passenger Sex
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df)
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# 4. Count Plot of Passenger Survival
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Count of Survivors (1: Survived, 0: Not Survived)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# 5. Box Plot of Fare by Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Class')
plt.xlabel('Class (Pclass)')
plt.ylabel('Fare')
plt.show()

# 6. Swarm Plot of Fare by Class
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Class (Swarm Plot)')
plt.xlabel('Class (Pclass)')
plt.ylabel('Fare')
plt.show()

# 7. Pair Plot of Numerical Columns
sns.pairplot(df[['Age', 'Fare', 'Parch', 'SibSp']], diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Columns')
plt.show()

# 8. Heatmap of Correlations
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# 9. Violin Plot of Age by Sex
plt.figure(figsize=(8, 6))
sns.violinplot(x='Sex', y='Age', data=df)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()

# 10. Bar Plot of Embarked Locations
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df)
plt.title('Count of Passengers by Embarked Location')
plt.xlabel('Embarked Location')
plt.ylabel('Count')
plt.show()
```
